```python
import numpy as np
import matplotlib.pyplot as plt
from tests.generate_wfdb import get_records, extract_signal_and_freq
from consensus_peaks import consensus_detect
```

```python
record_generator = get_records()
```

```python
%matplotlib widget
record = next(record_generator)
signal, freq_hz = extract_signal_and_freq(record)
signal = signal[:10_000]
peaks = consensus_detect(signal, freq_hz)

time = np.arange(0, len(signal)) / freq_hz
plt.plot(time, signal, label='ECG Signal')
peak_times = time[peaks]
plt.vlines(x=peak_times, ymin=np.min(signal), ymax=np.max(signal),
           linestyle='--', linewidth=0.5, label=f'Consensus Peaks')
```
